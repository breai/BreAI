```

import pandas as pd
import matplotlib.pyplot as plt
# Time [s], RESP, PLETH, V, AVR, II
bidmc1 = pd.read_csv("https://physionet.org/physiobank/database/bidmc/bidmc_csv/bidmc_01_Signals.csv", skipinitialspace=True , usecols=['Time [s]', 'RESP','PLETH'])
time = bidmc1['Time [s]']
resp = bidmc1.RESP
pleth = bidmc1['PLETH']
plt.plot(time, resp, label="rr")
plt.plot(time, pleth, label="ppg")
plt.plot()
# plt.plot(range(100))

plt.xlabel("time (s)")
plt.ylabel("y axis")
plt.title("RR vs. PPG")
plt.legend()
plt.show()
```
